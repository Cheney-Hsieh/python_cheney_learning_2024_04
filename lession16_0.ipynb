```python
import matplotlib.pyplot as plt
vendor_labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
my_colors =['Yellow','Green','Red','Blue']
my_explode =(0.2,0,0,0)


figure = plt.figure(figsize=(5, 5), dpi=100, facecolor='#ADD8E6')
axes = figure.add_subplot(1, 1, 1)
axes.pie(values, explode=my_explode, labels=vendor_labels, autopct='%1.1f%%',
        shadow=True, startangle=180,colors = my_colors)
axes.axis('equal')  
axes.set_title('Matplotlib Pie Exercise',color ='Red')
plt.show()
```
