list comprehension --> 快速建立list
pandas 可以補足numpy的不足,numpy無法加入欄位名稱跟索引


```python
import numpy as np
import pandas as pd
array_2d = np.random.randint(50,100,size=(50,5))
df1 = pd.DataFrame(array_2d,
             columns=['國文','英文','數學','地理','社會'],
             index = [f'{num}號' for num in range(1,51)]
             )
df1
```

```python
df1.values
```

```python
## dataFrame[欄位名稱] subscript
## DataFrame 實際是由多個 Series 組合起來的, 但是Series 沒有欄位名稱
type(df1['國文'])
```

```python
df1['國文'],df1['英文']
```

```python
##Series的選取,使用subscript
#Series[數值的index]

df1['國文'][:5] ## 盡量不要這麼用
df1['國文']['1號'] ## 正統用法
df1['國文'][['1號','2號','3號']] ## 正統用法
```

```python
##dataFrame 的 subscripts
#dataFrame[欄位名稱]
df1[['國文','英文']]
```

```python
##dataFrame 列的選取
## loc[字串] ,loc is property
#iloc[數值]
```

```python
#選取一號
df1.loc['1號']
print(type(df1.loc['1號']))
```

```python
#選取一號,二號,三號
df1.loc[['1號','2號','3號']]
print(type(df1.loc[['1號','2號','3號']]))
```

```python
## iloc 則輸入號碼　, loc 必須輸入字串
df1.iloc[-5:]
```
