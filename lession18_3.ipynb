```python
import numpy as np
import pandas as pd
scores = pd.DataFrame(np.random.randint(50,101,size=(50,5)),
             columns=['國文','英文','數學','地理','社會'],
             index = [f'{num}' for num in range(1,51)]
            )
scores
```

```python
## apply() 會一次%傳回一個Series 
def abc(s):
    return s.max()
scores.apply(abc)
```

```python
def xyz(s):
    return pd.Series([s.min(),s.max(),s.mean()],index=['最低分','最高分','平均'])
scores.apply(xyz)
```

```python
scores.index = scores.index.map(lambda i:f'{i}號')
scores
```

```python
def abc1(item):
    return pd.Series([item.sum(),item.mean()],index=['總分','平均'])
scores.apply(abc1,axis=1)
```

```python
one_student = scores.iloc[1]
less_student =one_student[one_student<60]
less_student
##less_student.index.values.tolist()
```

```python
less_student.count()
##less_student.index.values.tolist()
",".join(less_student.index.values.tolist())
```

```python
def abc2(item):
    less_student =item[item<60]
    less_count= less_student.count()
    less_str = ",".join(less_student.index.values.tolist())
    return pd.Series([item.sum(),item.mean(),less_count,less_str],index=['總分','平均','不及格的科目數','不及格科目'])
scores.apply(abc2,axis=1)
```
