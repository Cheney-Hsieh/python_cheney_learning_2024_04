```python
import pandas as pd
import pandas_datareader.data as pdr
import yfinance as yf
yf.pdr_override()
all_data = {item:pdr.get_data_yahoo(item) for item in ['2330.TW','2303.TW','2454.TW','2317.TW']}
stock_price= pd.DataFrame({key:df['Adj Close'] for key,df in all_data.items()})
stock_price.columns= ['台積電','聯電','聯發科','鴻海']
stock_price.iloc[-20:]
```

```python
df1 = stock_price.dropna()
df1.info()
df1
```
