```python
import numpy as np
import pandas as pd
array_2d = np.random.randint(50,101,size=(50,5))
df1 = pd.DataFrame(array_2d,
             columns=['國文','英文','數學','地理','社會'],
             index = [f'{num}號' for num in range(1,51)]
             )
df1
```

```python
total_score = df1['國文'] + df1['英文'] + df1['數學'] + df1['地理'] + df1['社會']
print(total_score)
```

```python
average_score = total_score /5
print(average_score)
```

```python
df1['總分'] = total_score
df1['平均'] = average_score
df1
```

```python
#Series 使用比較運算子
df1['總分'] >=400
```

```python
##dataFrame 使用subscripts, 搜索資料
#dataFrame[bool Series]
df1[df1['總分'] >=400]
```

```python
df1[(df1['國文'] >=90) & (df1['總分'] >= 400)]
```
